```python
print(len(catalog))
catalog = catalog[(catalog['R'] < 1000) & (catalog['lmass'] >= 10.2) ]
print(len(catalog))
```

```python
f1 = catalog[(catalog['Redshift'] >  0) & (catalog['z_peak'] >   0) & (abs(((catalog['z_peak'] - catalog['Redshift']))/(1+catalog['Redshift'] ))< 0.08)]
len(f1)
```

```python
f2  =  catalog[(catalog['Redshift'] <=  0) & (catalog['z_peak'] >  0)  ]
len(f2)
```

```python
id = []
mp = []
n_g = []
for i in range(len(f2)):
    id.append(f2[i])
    for j in range(len(f1)):
        if (abs(f1[j]['lmass'] - f2[i]['lmass']) <= 0.1 ):
           # print(abs(f1[j]['lmass'] - f2[i]['lmass'])  )
            mp.append(f1[j])
    n_g.append(mp)
    mp = []
```

```python
import math

sorted_5 = []

sort_l = []
for i in range(len(n_g)):
    for j in range(len(n_g[i])):
        sort_l.append((math.sqrt( ( n_g[i][j]['ra'] - id[i]['ra'])**2 + ( n_g[i][j]['dec'] - id[i]['dec'])**2  ), n_g[i][j]))
    sort_l = sorted(sort_l, key=lambda x: x[0])
    sorted_5.append(sort_l[0:5])
    sort_l=[]
```

```python
corr_i = []

 

for i in range(len(sorted_5)):
    sc = 0; si = 0; fp = 0; fn=0
    for j in range(len(sorted_5[i])):
       # if len(sorted_5[i])!=5  :
        #    print(len(sorted_5[i]))
         #
         #    continue
        if ( (((abs(sorted_5[i][j][1]['z_peak'] - sorted_5[i][j][1]['cluster_redshift']))) <= 0.08)  and  (((abs(sorted_5[i][j][1]['Redshift'] - sorted_5[i][j][1]['cluster_redshift'])) <= 0.02))):
            sc = sc+1
        if ((((abs(sorted_5[i][j][1]['z_peak'] - sorted_5[i][j][1]['cluster_redshift']))) <= 0.08)  and  ~(((abs(sorted_5[i][j][1]['Redshift'] - sorted_5[i][j][1]['cluster_redshift'])) <= 0.02))):
            fp = fp+1
        if ( ~(((abs(sorted_5[i][j][1]['z_peak'] - sorted_5[i][j][1]['cluster_redshift']))) <= 0.08) and  (((abs(sorted_5[i][j][1]['Redshift'] - sorted_5[i][j][1]['cluster_redshift'])) <= 0.02))):
            fn = fn+1
        else:
            si = si+1
    print(sc,fp,fn)

    res = 0
    if (sc+fp) == 0:
        res  = 0
    else: 
        res = (sc+fn)/(sc+fp) 
    corr_i.append((res, id[i] ))
    

```
